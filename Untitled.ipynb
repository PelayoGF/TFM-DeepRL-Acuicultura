```python
import pandas as pd
import numpy as np

# Cargar los datos reales
df = pd.read_csv("temperatures_real.csv", sep=";")

# Renombrar columnas para simplificar
df.rename(columns={"Fecha (aa/mm/dd)": "date"}, inplace=True)

# Convertir a datetime
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# Ordenar por fecha y quitar fechas inválidas
df = df.sort_values('date').dropna(subset=['date'])

# Establecer el índice como la fecha
df.set_index('date', inplace=True)

# Detectar la columna de temperatura (suponemos que es la única numérica si no hay encabezado raro)
temp_col = df.select_dtypes(include=[np.number]).columns[0]

# Interpolación avanzada (spline si hay suficientes datos, si no, fallback a time)
if df[temp_col].notna().sum() >= 4:  # spline requiere al menos 4 puntos
    df[temp_col] = df[temp_col].interpolate(method='spline', order=3)
else:
    df[temp_col] = df[temp_col].interpolate(method='time')

# Calcular la media semanal (usando semanas que terminan en lunes para compatibilidad)
weekly_df = df.resample('W-MON').mean().reset_index()

# Calcular el número de semana desde el 1 de enero de 2018
start_date = pd.to_datetime("2018-01-01")
weekly_df['week'] = ((weekly_df['date'] - start_date).dt.days // 7) + 1

# Reordenar columnas
weekly_df = weekly_df[['week', 'date', temp_col]]

# Guardar resultado
weekly_df.to_csv("temperatures_weekly.csv", index=False)

print(weekly_df.head())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# === PARTE 1: Procesar los datos reales diarios ===

# Cargar los datos reales
df = pd.read_csv("temperatures_real.csv", sep=";")

# Renombrar columnas para simplificar
df.rename(columns={"Fecha (aa/mm/dd)": "date"}, inplace=True)

# Convertir a datetime
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# Ordenar por fecha y quitar fechas inválidas
df = df.sort_values('date').dropna(subset=['date'])

# Establecer el índice como la fecha
df.set_index('date', inplace=True)

# Detectar la columna de temperatura (única numérica)
temp_col = df.select_dtypes(include=[np.number]).columns[0]

# Interpolación avanzada (spline si hay suficientes datos, sino temporal)
if df[temp_col].notna().sum() >= 4:
    df[temp_col] = df[temp_col].interpolate(method='spline', order=3)
else:
    df[temp_col] = df[temp_col].interpolate(method='time')

# Calcular la media semanal (agrupadas por lunes)
weekly_df = df.resample('W-MON').mean().reset_index()

# Calcular número de semana desde el 1 de enero de 2018
start_date = pd.to_datetime("2018-01-01")
weekly_df['week'] = ((weekly_df['date'] - start_date).dt.days // 7) + 1

# Limitar a las primeras 364 semanas
weekly_df = weekly_df[weekly_df['week'].between(1, 364)]

# Reordenar columnas
weekly_df = weekly_df[['week', 'date', temp_col]]

# Guardar resultado intermedio
weekly_df.to_csv("temperatures_weekly.csv", index=False)

# === PARTE 2: Interpolar faltantes y limpiar ===

df = pd.read_csv("temperatures_weekly.csv")

# Verificar y rellenar nulos
temp_col = df.select_dtypes(include=[np.number]).columns.difference(['week']).tolist()[0]
if df[temp_col].isna().sum() > 0:
    df[temp_col] = df[temp_col].interpolate(method='linear', limit_direction='both')

# Eliminar columna de fecha
df = df.drop(columns=['date'])

# Renombrar columnas
df = df.rename(columns={
    'week': 'Week',
    temp_col: 'TEMPERATURE'
})

# Guardar dataset final
df.to_csv("temperatures.csv", index=False)

# === PARTE 3: Graficar ===

# Cargar los datos procesados
df = pd.read_csv("temperatures.csv")

plt.figure(figsize=(14, 6))
plt.plot(df['Week'], df['TEMPERATURE'], marker='o', linestyle='-', color='royalblue')


plt.xlabel("Semana", fontsize=12)
plt.ylabel("Temperatura (°C)", fontsize=12)

plt.xticks(range(0, 365, 26))  # Etiquetas cada medio año
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
from scipy.stats import t

# -------------------------------------------------------------------------------------
# PROCESAMIENTO DE RECOMPENSAS CON MEDIAS MÓVILES DE 100 EPISODIOS
# -------------------------------------------------------------------------------------

window_size = 100  # Tamaño de la media móvil

# Paso 1: Determinar la longitud mínima común entre todos los runs
min_length = min(len(x) for x in all_moving_averages)

# Paso 2: Recortar todos los arrays a la longitud mínima
trimmed_rewards = np.array([x[:min_length] for x in all_moving_averages])  # Shape: (num_runs, min_length)

# Paso 3: Calcular medias móviles para cada ejecución individual
moving_averages = []
for run_rewards in trimmed_rewards:
    mov_avg = np.convolve(run_rewards, np.ones(window_size) / window_size, mode='valid')
    moving_averages.append(mov_avg)

# Paso 4: Convertir lista de medias móviles a array 2D (num_runs, num_windows)
moving_averages = np.array(moving_averages)

# Paso 5: Calcular estadísticos por ventana
mean_mov = np.mean(moving_averages, axis=0)
std_mov = np.std(moving_averages, axis=0, ddof=1)
stderr_mov = std_mov / np.sqrt(len(moving_averages))

# Paso 6: Calcular intervalo de confianza 95%
confidence_level = 0.95
t_value = t.ppf((1 + confidence_level) / 2, df=len(moving_averages) - 1)
ci_mov = t_value * stderr_mov

# Paso 7: Eje X: episodios centrados en cada ventana
episode_indices = np.arange(window_size // 2, window_size // 2 + len(mean_mov))

# Paso 8: Graficar resultados
plt.figure(figsize=(10, 6))
plt.plot(episode_indices, mean_mov, label="Media móvil (100 episodios)", color="blue")
plt.fill_between(episode_indices, mean_mov - ci_mov, mean_mov + ci_mov, color="blue", alpha=0.3,
                 label="Intervalo de confianza 95%")
plt.xlabel("Episodio")
plt.ylabel("Recompensa media")
plt.xlim(1, 6000)     # Limita el eje X de 0 a 1000 episodios
plt.grid(False)
plt.tight_layout()
plt.show()
```
